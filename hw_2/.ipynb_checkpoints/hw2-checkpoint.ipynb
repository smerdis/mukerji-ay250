### Question 0

I will analyze the figure on page 120 of Hubel and Wiesel's 1962 paper. The link: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1359523/?page=15 should take you to it. Alternatively, here's a link to a screenshot: https://www.dropbox.com/s/suqqd26l046jm62/Screenshot%202018-02-12%2011.14.52.png?dl=0

This figure illustrates the firing pattern of a particular neuron in the primary visusal cortex of a cat. ***The major point it seeks to convey is that this cell is orientation-selective***: it responds maximally to a horizontal black bar placed within its receptive field, and not at all to a bar which has been rotated away from horizontal. In my opinion, it conveys this well, because I didn't even need to read the caption to understand this takeaway.

The only major critique I have of this figure is that the diagrams on the left showing the stimulus position contain dashed lines of seemingly no significance. The caption gives the size of the visual field, and one could interpret these lines in light of that, but in my opinion they clutter up the figure a bit. Removing them would still convey the same story.

```python
from bokeh.models import ColumnDataSource, Span
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

df = pd.read_csv('master_df_bloom_figure.csv')

from bokeh.plotting import figure, show, output_file, output_notebook

## show inline in the notebook
output_notebook()

p = figure(title="Question 2", toolbar_location="above",
           x_axis_label="year", y_axis_label="Gflops per Watt", 
           y_axis_type='log', y_range=[1e-2, 1e6], plot_width=600, plot_height=600)

groups = df.groupby(['type'])
for gv, g in groups:
    source_t = ColumnDataSource(g.groupby('year'))
    print(source_t.column_names)
    p.vbar(x='year', width=0.5, bottom="gpw_25%", top="gpw_75%", source=source_t)
    p.vbar(x='year', width=0.5, bottom="gpw_50%", top="gpw_50%", source=source_t)
    #p.scatter(x='year', y="gpw_count", size=3,source=source_t)

hline = Span(location=1e5, dimension='width', line_color='black', line_width=3)
p.renderers.extend([hline])
show(p)
```

### Question 2

```python
data_files = {"google":"hw_2_data/google_data.txt", "ny":"hw_2_data/ny_temps.txt",
              "yahoo":"hw_2_data/yahoo_data.txt"}
for k, v in data_files.items():
    print(k, v)

# Read the files, replace the non-date column name with the key from above
google_df, ny_df, yahoo_df = (pd.read_csv(f, sep='\t', header=0, names=['MJD', k]) for k,f in data_files.items())
```

```python
# outer joins preserve all elements of both inputs
combined_df = pd.merge(google_df, ny_df, how="outer")
combined_df = pd.merge(combined_df, yahoo_df, how="outer")
```

```python
def plot_graph(df):
    f, ax = plt.subplots(figsize=(10,10))

    # x value - date
    x = df.MJD
    # second y axis for temperature
    ax2 = ax.twinx()
    
    # Plot each with a label (for a legend)
    ax.plot(x, df.google, label='Google Stock Value', color='blue')
    ax2.plot(x, df.ny, label='NY Mon. High Temp', linestyle='--', color='red')
    ax.plot(x, df.yahoo, label='Yahoo! Stock Value', color='purple')
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1000))
    ax.xaxis.set_minor_locator(ticker.MultipleLocator(200))

    ax.yaxis.set_major_locator(ticker.MultipleLocator(100))
    ax.yaxis.set_minor_locator(ticker.MultipleLocator(20))

    ax2.yaxis.set_major_locator(ticker.MultipleLocator(50))
    ax2.yaxis.set_minor_locator(ticker.MultipleLocator(10))
    ax2.set_ylim(-150, 100)

    lines, labels = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc="center left")

    # Set main text elements of the plot
    ax.set_title('New York Temperature, Google, and Yahoo!', fontsize=24)
    ax.set_xlabel('Date (MJD)',fontsize=16)
    ax.set_ylabel('Value (Dollars)',fontsize=16)
    ax2.set_ylabel('Temperature (°F)', fontsize=16)
```

```python
plot_graph(combined_df)
```
